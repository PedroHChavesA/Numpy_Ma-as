Os dados referentes a laranja vão até a linha 4999 e os referentes à toranja iniciam na linha 5000 do arquivo.

Após fazer a seleção de dados, importe a biblioteca matplotlib e crie um gráfico para a laranja e para a toranja do peso pelo diâmetro.

```python
laranja_diamentro = dados[:5000,0]
laranja_peso = dados[:5000,1]

toranja_diamentro = dados[5000:, 0]
toranja_peso = dados[5000:, 1]
```

```python
import matplotlib.pyplot as plt
plt.plot(laranja_diamentro, laranja_peso)
plt.plot(toranja_diamentro, toranja_peso)
```

```python
X = laranja_diamentro

Y = laranja_peso

n = np.size(X)

```

```python
a = (n * np.sum(X * Y) - np.sum(X) * np.sum(Y)) / (n * np.sum(X**2) - np.sum(X)**2)

b = np.mean(Y) - a * np.mean(X)

y = a*X+b

x = laranja_diamentro

np.linalg.norm(laranja_diamentro - y)
```

```python
plt.plot(laranja_diamentro, laranja_peso)
plt.plot(x, y)
```
